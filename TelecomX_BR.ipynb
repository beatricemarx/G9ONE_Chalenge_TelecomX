#📌 Extracão

```python
import requests
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'

response = requests.get(url)

response

```

```python
#transforma em um arquivo python - formato dicionário

data = response.json()

```

#🔧 Transformação

```python
#transformar em dataframe

dados = pd.DataFrame(data)

dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
def normalize_and_concat(df, column_name):
    normalized_df = pd.json_normalize(df[column_name])

    # NÃO adicionar prefixo nas colunas
    # normalized_df.columns = [f"{column_name}_{col}" for col in normalized_df.columns]

    df = pd.concat([df.drop(columns=[column_name]), normalized_df], axis=1)
    return df

# Aplicar a normalização para cada coluna com dicionário aninhado
for col in colunas[2:]:  # ['customer', 'phone', 'internet', 'account']
    dados = normalize_and_concat(dados, col)

# Visualizar resultado
dados
```

```python
#converter espaços em branco para NA e depois converter a coluna Charges.Total para numérica
dados['Charges.Total'] = dados['Charges.Total'].replace(' ', pd.NA)
dados['Charges.Total'] = pd.to_numeric(dados['Charges.Total'], errors='coerce')
dados
```
