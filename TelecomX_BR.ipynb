```python
def normalize_and_concat(df, column_name):
    normalized_df = pd.json_normalize(df[column_name])

    # NÃO adicionar prefixo nas colunas
    # normalized_df.columns = [f"{column_name}_{col}" for col in normalized_df.columns]

    df = pd.concat([df.drop(columns=[column_name]), normalized_df], axis=1)
    return df

# Aplicar a normalização para cada coluna com dicionário aninhado
for col in colunas[2:]:  # ['customer', 'phone', 'internet', 'account']
    dados = normalize_and_concat(dados, col)

# Visualizar resultado
dados
```

```python
media_mensal = (
    dados
    .groupby("Churn")["Cobrança Mensal"]
    .mean()
    .round(2)
)


import matplotlib.pyplot as plt

# 🎨 Paleta personalizada
cor_fundo = "#EEE3DC"
cor_texto = "#3A2D32"
cores = ["#A2AD91", "#D30C7B"]  # Não | Sim

fig, ax = plt.subplots(figsize=(6,5))

fig.patch.set_facecolor(cor_fundo)
ax.set_facecolor(cor_fundo)

bars = ax.bar(
    media_mensal.index,
    media_mensal.values,
    color=cores,
    edgecolor="white"
)

ax.set_title(
    "Média da Cobrança Mensal por Churn",
    fontsize=15,
    weight="bold",
    color=cor_texto
)

ax.set_ylabel("Cobrança Mensal Média (R$)", color=cor_texto)
ax.set_xlabel("")
ax.tick_params(colors=cor_texto)

# 🔥 Inserir valor acima das barras
for bar in bars:
    altura = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,
        altura,
        f"R$ {altura:.2f}",
        ha="center",
        va="bottom",
        fontsize=11,
        weight="bold",
        color=cor_texto
    )

plt.tight_layout()
plt.show()
```

#📄Relatorio Final

O objtivo do projeto "Churn de Clientes" foi analisar os dados para entender os motivos do alto índice de cancelamentos que a empresa Telecom X está sofrendo.

Os dados foram coletados, tratados e analisados usando o Python e suas principais bibliotecas para extrair insights. O primeiro passo foi a exportação dos dados no formato json, e depois conversão para um datrafame. Foi necessário normalizar os dados, uma vez que algumas colunas estavam em formato de dicionário. Depois disso foi feita a limpeza dos dados, tirando linhas com valores nulos e ou em branco, ajuste dos formatos das colunas que tinham dados numéricos mas estavam como string, e padronização dos títulos e dados que foram traduzidos para o português.

Então, o próximo passo foi a exploração dos dados, a fim de conhecer s variáveis e entender o problema do churn elevado, que está em torno de 26%.

O primeiro ponto de análise foi entender o público alvo da empresa. A análise feita mostra que a maior parte dos clientes da empresa são pessoas com menos de 60 anos, sem filhos e solteiros e estes são os mais propensos a cancelar o serviço. Assim recomendo que sejam feitas campanhas com foco neste perfil de consumidor.

Outro ponto visto foi que há uma grande tendência de cancelamento de serviços no início do contrato, e que a maior parte dos cancelamento está com os clientes que tem contratos mensais. Assim, campanhas que ofereçam um estímulo financeiro para fechar contratos anuais ou com 2 anos podem ajudar a reduzir os cancelamentos.
