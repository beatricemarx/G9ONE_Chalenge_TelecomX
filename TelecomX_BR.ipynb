```python
def normalize_and_concat(df, column_name):
    normalized_df = pd.json_normalize(df[column_name])

    # NÃO adicionar prefixo nas colunas
    # normalized_df.columns = [f"{column_name}_{col}" for col in normalized_df.columns]

    df = pd.concat([df.drop(columns=[column_name]), normalized_df], axis=1)
    return df

# Aplicar a normalização para cada coluna com dicionário aninhado
for col in colunas[2:]:  # ['customer', 'phone', 'internet', 'account']
    dados = normalize_and_concat(dados, col)

# Visualizar resultado
dados
```

```python
#gráfico em que o tamanho da barra representa o percentual em relação ao total de clientes.

import pandas as pd
import matplotlib.pyplot as plt

# Criar cópia
dados_plot = dados.copy()

# 🔹 Renomeações


dados_plot["Idoso"] = dados_plot["Idoso"].replace({
    "Não": "<60",
    "Sim": "60+",
    0: "<60",
    1: "60+"
})

dados_plot["Parceiro"] = dados_plot["Parceiro"].replace({
    "Sim": "Comprometidos",
    "Não": "Solteiros"
})

dados_plot["Dependentes"] = dados_plot["Dependentes"].replace({
    "Sim": "Com Filhos",
    "Não": "Sem Filhos"
})

colunas = ["Gênero", "Idoso", "Parceiro", "Dependentes"]

tabelas = []
total_clientes = len(dados_plot)

for col in colunas:

    temp = (
        dados_plot
        .groupby([col, "Churn"])
        .size()
        .unstack(fill_value=0)
    )

    # 🔥 Aqui está a mudança principal
    temp = (temp / total_clientes) * 100

    temp["Variavel"] = col
    temp["Categoria"] = temp.index
    tabelas.append(temp.reset_index(drop=True))

df_plot = pd.concat(tabelas)

df_plot["Grupo"] = df_plot["Variavel"] + " - " + df_plot["Categoria"]

#df_plot["Grupo"] = df_plot["Categoria"]

df_plot = df_plot.set_index("Grupo")[["Não", "Sim"]]





#Visualização do gráfico

# 🎨 Cores da paleta
cor_fundo = "#EEE3DC"
cor_texto = "#3A2D32"
cores = ["#A2AD91", "#D30C7B"]  # No | Yes

fig, ax = plt.subplots(figsize=(10,6))

# Fundo personalizado
fig.patch.set_facecolor(cor_fundo)
ax.set_facecolor(cor_fundo)

df_plot.plot(
    kind="barh",
    stacked=True,
    color=cores,
    edgecolor="white",
    ax=ax
)

ax.set_title(
    "Distribuição Percentual Real sobre o Total de Clientes",
    fontsize=15,
    weight="bold",
    color=cor_texto
)

ax.set_xlabel("Percentual sobre o total (%)", color=cor_texto)
ax.set_ylabel("")

ax.tick_params(colors=cor_texto)

ax.legend(
    title="Churn",
    frameon=False
)

plt.tight_layout()
plt.show()
```
