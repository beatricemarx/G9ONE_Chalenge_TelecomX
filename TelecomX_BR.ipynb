```python
def normalize_and_concat(df, column_name):
    normalized_df = pd.json_normalize(df[column_name])

    # NÃO adicionar prefixo nas colunas
    # normalized_df.columns = [f"{column_name}_{col}" for col in normalized_df.columns]

    df = pd.concat([df.drop(columns=[column_name]), normalized_df], axis=1)
    return df

# Aplicar a normalização para cada coluna com dicionário aninhado
for col in colunas[2:]:  # ['customer', 'phone', 'internet', 'account']
    dados = normalize_and_concat(dados, col)

# Visualizar resultado
dados
```

# Distribuição da Evasão

Neste passo, o objetivo é entender como a variável "churn" (evasão) está distribuída entre os clientes. Utilize gráficos para visualizar a proporção de clientes que permaneceram e os que saíram.

```python
#agrupar por gender

dados.groupby("gender").size()
```

```python
#agrupar por churn

dados.groupby("Churn").size()
```

```python
#calcular a % de Churn

evasao_percentual = dados["Churn"].value_counts(normalize=True)*100

evasao_percentual

```

```python
import matplotlib.pyplot as plt

# Cores da paleta
cor_fundo = "#EEE3DC"
cor_texto = "#3A2D32"
cores = ["#A2AD91", "#D30C7B"]  # No | Yes

fig, ax = plt.subplots(figsize=(5,5))

# Fundo da figura
fig.patch.set_facecolor(cor_fundo)
ax.set_facecolor(cor_fundo)

wedges, texts, autotexts = ax.pie(
    evasao_percentual.values,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.78,
    colors=cores,
    wedgeprops={"width": 0.40, "edgecolor": cor_fundo},
    textprops={"fontsize": 12, "weight": "bold", "color": cor_texto}
)

# Ajustar tamanho dos percentuais
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_weight("bold")
    autotext.set_color(cor_texto)

ax.set_title(
    "Distribuição de Evasão de Clientes",
    fontsize=16,
    weight="bold",
    color=cor_texto,
    pad=20
)

ax.axis("equal")

# Texto central
ax.text(
    0, 0,
    f"{len(dados)}\nClientes",
    ha="center",
    va="center",
    fontsize=14,
    weight="bold",
    color=cor_texto
)

# Legenda minimalista
ax.legend(
    wedges,
    evasao_percentual.index,
    title="Evasão",
    loc="lower center",
    bbox_to_anchor=(0.5, -0.1),
    ncol=2,
    frameon=False
)

plt.tight_layout()
plt.show()
```

```python
#verificar a distribuição por tipo de gender - distribuição igual entre os gêneros

def perfil_por_churn(coluna):
    tabela = (
        dados
        .groupby("Churn")[coluna]
        .value_counts(normalize=True)
        .unstack()
        * 100
    )
    return tabela.round(2)

perfil_gender = perfil_por_churn("gender")


#mudar o nome das colunas de valor

percentual_gender = perfil_gender.rename(columns={"Female": "Mulheres", "Male": "Homens"})
percentual_gender
```

```python
#verificar a distribuição por tipo de SeniorCitizenp  - O churn é mais alto entre os jovens

perfil_seniorCitizen = perfil_por_churn("SeniorCitizen")


#mudar o nome das colunas de valor

percentual_seniorCitizen = perfil_seniorCitizen.rename(columns={"No": "Jovens", "Yes": "Idosos"})
percentual_seniorCitizen
```

```python
#verificar a distribuição por tipo de Partner - o churn é mais alto entre os solteiros

perfil_partner = perfil_por_churn("Partner")


#mudar o nome das colunas de valor

percentual_partner = perfil_partner.rename(columns={"No": "Solteiros", "Yes": "Comprometidos"})
percentual_partner
```

```python
#verificar a distribuição por tipo de Partner - o churn é mais alto entre os sem filhos

perfil_dependents = perfil_por_churn("Dependents")

#mudar o nome das colunas de valor

percentual_dependents = perfil_dependents.rename(columns={"No": "Sem Filhos", "Yes": "Com Filhos"})
percentual_dependents
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar uma cópia para não alterar o dataframe original
dados_plot = dados.copy()

# 🔹 Renomeações
dados_plot["gender"] = dados_plot["gender"].replace({
    "Male": "Homem",
    "Female": "Mulher"
})

dados_plot["SeniorCitizen"] = dados_plot["SeniorCitizen"].replace({
    "No": "Jovens",
    "Yes": "Idosos",
    0: "Jovens",
    1: "Idosos"
})

dados_plot["Partner"] = dados_plot["Partner"].replace({
    "Yes": "Comprometidos",
    "No": "Solteiros"
})

dados_plot["Dependents"] = dados_plot["Dependents"].replace({
    "Yes": "Com Filhos",
    "No": "Sem Filhos"
})

# Lista das colunas que queremos analisar
colunas = ["gender", "SeniorCitizen", "Partner", "Dependents"]

# Criar tabela consolidada
tabelas = []

for col in colunas:
    temp = (
        dados_plot
        .groupby(col)["Churn"]
        .value_counts(normalize=True)
        .unstack()
        * 100
    )

    temp["Variavel"] = col
    temp["Categoria"] = temp.index
    tabelas.append(temp.reset_index(drop=True))

# Unir tudo
df_plot = pd.concat(tabelas)

# Criar rótulo combinado
df_plot["Grupo"] = df_plot["Variavel"] + " - " + df_plot["Categoria"]

df_plot = df_plot.set_index("Grupo")[["No", "Yes"]]



# Visualização


# Paleta elegante
cor_fundo = "#EEE3DC"
cor_texto = "#3A2D32"
cores = ["#A2AD91", "#D30C7B"]  # No | Yes

fig, ax = plt.subplots(figsize=(10,6))

fig.patch.set_facecolor(cor_fundo)
ax.set_facecolor(cor_fundo)

df_plot.plot(
    kind="bar",
    stacked=True,
    color=cores,
    edgecolor="white",
    ax=ax
)

ax.set_title("Perfil dos Clientes por Categoria",
             fontsize=16, weight="bold", color=cor_texto)

ax.set_ylabel("Percentual (%)", color=cor_texto)
ax.set_xlabel("")
ax.tick_params(axis="x", rotation=45)

ax.legend(title="Churn", frameon=False)

plt.tight_layout()
plt.show()
```

```python

#gráfico em que o tamanho da barra representa o percentual em relação ao total de clientes.

import pandas as pd
import matplotlib.pyplot as plt

# Criar cópia
dados_plot = dados.copy()

# 🔹 Renomeações
dados_plot["gender"] = dados_plot["gender"].replace({
    "Male": "Homem",
    "Female": "Mulher"
})

dados_plot["SeniorCitizen"] = dados_plot["SeniorCitizen"].replace({
    "No": "Jovens",
    "Yes": "Idosos",
    0: "Jovens",
    1: "Idosos"
})

dados_plot["Partner"] = dados_plot["Partner"].replace({
    "Yes": "Comprometidos",
    "No": "Solteiros"
})

dados_plot["Dependents"] = dados_plot["Dependents"].replace({
    "Yes": "Com Filhos",
    "No": "Sem Filhos"
})

colunas = ["gender", "SeniorCitizen", "Partner", "Dependents"]

tabelas = []
total_clientes = len(dados_plot)

for col in colunas:

    temp = (
        dados_plot
        .groupby([col, "Churn"])
        .size()
        .unstack(fill_value=0)
    )

    # 🔥 Aqui está a mudança principal
    temp = (temp / total_clientes) * 100

    temp["Variavel"] = col
    temp["Categoria"] = temp.index
    tabelas.append(temp.reset_index(drop=True))

df_plot = pd.concat(tabelas)

df_plot["Grupo"] = df_plot["Variavel"] + " - " + df_plot["Categoria"]

df_plot = df_plot.set_index("Grupo")[["No", "Yes"]]





#Visualização do gráfico

# 🎨 Cores da paleta
cor_fundo = "#EEE3DC"
cor_texto = "#3A2D32"
cores = ["#A2AD91", "#D30C7B"]  # No | Yes

fig, ax = plt.subplots(figsize=(10,6))

# Fundo personalizado
fig.patch.set_facecolor(cor_fundo)
ax.set_facecolor(cor_fundo)

df_plot.plot(
    kind="barh",
    stacked=True,
    color=cores,
    edgecolor="white",
    ax=ax
)

ax.set_title(
    "Distribuição Percentual Real sobre o Total de Clientes",
    fontsize=15,
    weight="bold",
    color=cor_texto
)

ax.set_xlabel("Percentual sobre o total (%)", color=cor_texto)
ax.set_ylabel("")

ax.tick_params(colors=cor_texto)

ax.legend(
    title="Churn",
    frameon=False
)

plt.tight_layout()
plt.show()
```
